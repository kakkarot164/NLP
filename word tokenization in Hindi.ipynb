```python
import nltk
from nltk.tokenize import word_tokenize

hindi_text = """प्राकृतिक भाषा सीखना बहुत रोचक है।"""

# Tokenization
tokens = word_tokenize(hindi_text)

# Define stopwords for Hindi
stop_words = ["है", "हैं", "के", "का", "की", "को", "में", "और", "से", "होता", "होती", "होते"]

# Remove stopwords
filtered_tokens = [token for token in tokens if token not in stop_words]

print("Original Text: ", hindi_text)
print("Tokenized Text: ", tokens)
print("Filtered Text: ", filtered_tokens)
```
